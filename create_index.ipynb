```python
def createlink():
    linkout={}
    linkin={}
    ps=ParserCACM()
    ps.initFile("cacm/cacm.txt")
    doc=ps.nextDocument()
    while doc:
        linkout[doc.getId()]=set()
        links=doc.get('links').split(';')
        for link in links:
            if link!='':
                linkout[doc.getId()].add(link)
                if link not in linkin:
                    linkin[link]=set()    # value unique            
                linkin[link].add(doc.getId())                             
        doc=ps.nextDocument() 
    return linkin,linkout
```
